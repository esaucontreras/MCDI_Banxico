# CONVERSIÓN DE PDF A TXT

```python
import os
import pdfplumber
```

```python
# SE LEE DIRECTORIO Y ALMACENAN NOMBRES DE ARCHIVOS
archivos_1= os.listdir()
#archivos[5]
archivos = archivos_1[5:9]+archivos_1[11:-1]
```

```python
# COMO LOS ARCHIVOS DEL 2000 SE LEEN CON ERRORES
# SE LEEN LOS ARCHIVOS   A PARTIR DEL 2001
no_leidos = []

for i in range(0, len(archivos)):
    with pdfplumber.open(archivos[i]) as pdf:
        guardar = archivos[i][:-3]
        q = 0
        txt = ''
        while q<len(pdf.pages):
            for j in range(len(pdf.pages)):
                pagina = pdf.pages[j]
                extraccion = pagina.extract_text()
                txt = txt + " "+ extraccion
                #GUARDAR
                
                q+=1
        try:
            f = open('C:\\Users\\ESAU-C\\Documents\\MCDI INFOTEC\\Tesis\\Datos_txt\\'+guardar+'.txt',"w")
            f.write(txt)
            f.close()
                #print(q,len(pdf.pages) )
            print("El archivo pdf "+str(i)+" "+archivos[i]+" se ha convertido a texto.")     
        
        except:
            no_leidos.append(archivos[i])
            print("El archivo pdf "+str(i)+" "+archivos[i]+" no puede leerse.")
```
